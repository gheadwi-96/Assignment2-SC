### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import cluster
from sklearn.metrics import silhouette_score
```

### Read Dataset

```python
data = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/cluster_s1.csv')
data = data.drop('no', axis = 1)
data.head()
```

```python
%matplotlib inline
ax1 = data.plot.scatter(x='x', y='y', c='DarkBlue')
```

### Data Segmentation (Clustering)

```python
# TODO: Lakukan pencarian jumlah cluster terbaik berdasarkan data diatas
from sklearn.cluster import KMeans
K = list(range(2, 11))
silhouette_scores = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(data)
    silhouette = silhouette_score(data, cluster_labels)
    silhouette_scores.append(silhouette)

n_clust = K[silhouette_scores.index(max(silhouette_scores))]
n_clust, max(silhouette_scores)
```

```python
# TODO: Lakukan pemodelan dengan KMeans
kmeans = cluster.KMeans(n_clusters=n_clust, random_state=0)
kmeans.fit(data)
label = kmeans.labels_

score = silhouette_score(data, label)
print("silhouette score={:.5f}".format(score))
```

### Visualization: Cluster Result

```python
# TODO: Masukkan label hasil clustering ke dalam DataFrame

result = data.copy()
result['cluster'] = label
result.head()
```

```python
# TODO: Plot data hasil clustering dengan Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', hue='cluster', palette='muted', data=result)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
